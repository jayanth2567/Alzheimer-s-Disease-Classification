```python

bagg_finalvo=BaggingClassifier(estimator=model_finalvo,n_estimators=100,max_samples=0.7,max_features=1.0,bootstrap=True,bootstrap_features=False,random_state=42)

lgbm_final1=LGBMClassifier(
    bagging_fraction=1.0,
    bagging_freq=1,
    feature_fraction=0.8,
    learning_rate=0.05,
    max_depth=7,
    n_estimators=100,
    num_leaves=15,
    reg_alpha=1,
    reg_lambda=0,
    random_state=42
)

xgb_final = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss'
)

# 7. Create Voting Ensemble (Soft Vote averages probabilities)
voting_clf123 = VotingClassifier(
    estimators=[
        ('bagg', bagg_finalvo),
        ('lgbm', lgbm_final1),
        ('xgb',xgb_final)
    ],
    voting='hard'
)
voting_clf123.fit(x_train,y_train)
```

```python

bagg_finalvo=BaggingClassifier(estimator=model_finalvo,n_estimators=100,max_samples=0.7,max_features=1.0,bootstrap=True,bootstrap_features=False,random_state=42)

lgbm_final1=LGBMClassifier(
    bagging_fraction=1.0,
    bagging_freq=1,
    feature_fraction=0.8,
    learning_rate=0.05,
    max_depth=7,
    n_estimators=100,
    num_leaves=15,
    reg_alpha=1,
    reg_lambda=0,
    random_state=42
)

xgb_final = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss'
)
catboost151=CatBoostClassifier(bagging_temperature=0,border_count=128,depth=6,iterations=200,l2_leaf_reg=5,learning_rate=0.01,random_strength=1,random_state=42)
# 7. Create Voting Ensemble (Soft Vote averages probabilities)
stacking_clf = StackingClassifier(
    estimators=[
        ('bagg', bagg_finalvo),
        ('lgbm', lgbm_final1),
        ('xgb',xgb_final),
        ('cat',catboost151)
    ],
   final_estimator=LogisticRegression(), # Meta-model
    cv=5,  # Internal cross-validation to prevent leakage
    n_jobs=-13
)
stacking_clf.fit(x_train,y_train)
```
